## Предобработка данных

**Переименуем столбцы:**

```python
df = df.rename(columns={
    'studio': 'is_studio',
    'open_plan': 'is_open_plan',
    'cityCenters_nearest': 'city_centers_nearest',
    'parks_around3000': 'parks_around_3km',
    'ponds_around3000': 'ponds_around_3km'
})
```

**Заполним пропуски в данных:**

```python
# пропуски заполняем 0, так как имеется ввиду отсутствие балкона:
df.balcony = df.balcony.fillna(0)
```

```python
# удалим строки с пропусками в этих столбцах, так как пропущенных значений мало и их сложно коректно исправить:
df.dropna(subset=['floors_total', 'locality_name'], inplace=True)
```

```python
# найдем медианное значение высоты потолка,
# заполним пропуски этим значением:
median_ceiling_height = df.ceiling_height.median()
df.ceiling_height = df.ceiling_height.fillna(median_ceiling_height)
```

```python
# люди у кого в недвижимости апартаменты, обычно знают об этом, поэтому заполним пропуски False:
df.is_apartment = df.is_apartment.fillna(False)
```

```python
# заполним пропуски в living_area медианой по числу комнат:
for room in df.rooms.unique():
    median = df.loc[df['rooms'] == room, 'living_area'].median()
    df.loc[(df['living_area'].isna()) & (df['rooms'] == room), 'living_area'] = median
```

```python
# зная общую и жилую площадь, мы можем разностью заполнить kitchen_area:
df.kitchen_area = df.loc[df['kitchen_area'].isna(), 'kitchen_area'] = df.total_area - df.living_area
```

```python
# но важно учесть, что для квартир студий kitchen_area равна 0, а living_area равна total_area:
for flat_type in df.is_studio.unique():
    if flat_type == True:
        df.kitchen_area = 0
        df.living_area = df.total_area
```

```python
# для locality_name оставим только названия:
for name in ['посёлок ', 'городской ', 'имени ', 'городского типа ', 'городской посёлок ', 'станции ',
             'коттеджный ', 'деревня ', 'поселок городского типа ', 'при железнодорожной станции ',
             'поселок ', 'садовое товарищество ', 'село ', 'городской поселок ', 'поселок станции ',
             'садоводческое некоммерческое товарищество ', 'при железнодорожной ', 'Пансионат ', 'Форт ',
             'посёлок городского типа ', 'коттеджный поселок ', 'посёлок при железнодорожной станции ']:
    df['locality_name'] = df['locality_name'].str.replace(name, '')
```

```python
# узнаем сколько пропущено значений расстояния до центра для квартир из Санкт-Петербурга,
# так как эти данные понадобятся нам в исследовании:
print(len(df[(df['locality_name'] == 'Санкт-Петербург') & (df['city_centers_nearest'].isna())]))
round(len(df[(df['locality_name'] == 'Санкт-Петербург') & (df['city_centers_nearest'].isna())]) /
      len(df.query('locality_name == "Санкт-Петербург"')), 3)
```

```python
# поскольку количество пропусков очень мало, а их заполнение может быть трудоемким, просто удалим эти пропуски:
df = df.query('~(locality_name == "Санкт-Петербург" & city_centers_nearest.isna())').reset_index(drop=True)
```

```python
# пропуски в длительности размещения объявления (от публикации до снятия) заменим на медианное значение:
df.loc[df['days_exposition'].isna(), 'days_exposition'] = df['days_exposition'].median()
```

Пропуски в столбцах:

+ `airports_nearest` - расстояние до ближайшего аэропорта,
+ `parks_around_3km` - число парков в радиусе 3 км,
+ `parks_nearest` - расстояние до ближайшего парка,
+ `ponds_around_3km` - число водоемов в радиусе 3 км,
+ `ponds_nearest` - расстояние до ближайшего водоема,

мы можем оставить без изменений, так как они не будут использоваться в текущем исследовании.

**Изменим тип данных:**

```python
# изменим last_price на тип int для удобства просмотра:
df.last_price = df.last_price.astype('int')

# меняем тип данных на datetime, так как это время:
df.first_day_exposition = pd.to_datetime(df.first_day_exposition, format='%Y-%m-%dT%H:%M:%S')

# изменим количество этажей на int, поскольку это целое число:
df.floors_total = df.floors_total.astype('int')

# меняем тип данных на float, для площади кухни:
df.kitchen_area = df.kitchen_area.astype('float')

# меняем тип данных на int, так как количество балконов, это целое число:
df.balcony = df.balcony.astype('int')

# количество дней размещения объявления соответствует типу int:
df.days_exposition = df.days_exposition.astype('int')
```

## Расчёты и добавление результатов в таблицу

**Цена квадратного метра:**

```python
df['price_per_sqm'] = df['last_price'] / df['total_area']

# изменим price_per_sqm на тип int для удобства просмотра:
df.price_per_sqm = df.price_per_sqm.astype('int')
```

**День недели, месяц и год публикации объявления:**

```python
df['weekday'] = df['first_day_exposition'].dt.weekday
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year
```

**Этаж квартиры (первый, последний, другой):**

```python
# напишем функцию для категоризации этажей:
def floor_category(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'первый'
    elif floor == row['floors_total']:
        return 'последний'
    elif 1 < floor < row['floors_total']:
        return 'другой'
    
# применим написанную функцию к каждой строке dataframe  
df['floor_category'] = df.apply(floor_category, axis = 1)
```

**Соотношение жилой и общей площади, а также отношение площади кухни к общей:**

```python
df['living_area_ratio'] = df['living_area'] / df['total_area']
df['kitchen_area_ratio'] = df['kitchen_area'] / df['total_area']
```

## Исследовательский анализ данных

```python
# напишем функцию возвращающую границы 'усов' для конкретного признака:
def limit_whisker(column):
    info = df[column].describe()
    # межквартильный размах:
    iqr = info[6] - info[4]
    # левый и правый 'ус':
    left_whisker = round(info[4] - 1.5 * iqr, 2)
    right_whisker = round(info[6] + 1.5 * iqr, 2)
    # 'усы' не могут быть больше максимума и минимума:
    if left_whisker < info[3]:
        left_whisker = info[3]
    if right_whisker > info[7]:
        right_whisker = info[7]
    return [left_whisker, right_whisker]
```
